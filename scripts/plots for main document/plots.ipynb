```python
%pylab inline
import mpl_toolkits.mplot3d.axes3d as p3
```

```python
def Re(N): return 0.012 + N * 0.000026
def Rb(N): return N * (0.012 + 0.000026)

def Te(N,M,B): return B*2*Re(N)
def Tb(N,M,B): return B*Rb(M)
```

```python
dataE = array([[ Te(N,M,1) for N in range(100)] for M in range(100)])
dataB = array([[ Tb(N,M,1) for N in range(100)] for M in range(100)])
```

```python
fig = figure()
ax = p3.Axes3D(fig)

x,y = mgrid[0:100,0:100]
ax.plot_surface(y, x, dataE, cmap=cm.Blues)
ax.plot_surface(y, x, dataB, cmap=cm.Reds)

ax.set_xlabel('M')
ax.set_ylabel('N')
ax.set_zlabel('total time')
#ax.axis([0,10,0,10])
show()
```
