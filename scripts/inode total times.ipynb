```python
def R1(N): return 0.010 + N * 3e-5
def RN(N): return N * (0.010 + 3e-5)

def T1(N,M,B): return sum( 2*R1(N)  for i in range(B))
def TN(N,M,B): return sum( RN(M)    for i in range(B))
```

```python
data1 = array([[ T1(N,M,1) for N in range(100)] for M in range(100)])
dataN = array([[ TN(N,M,1) for N in range(100)] for M in range(100)])
```

```python
from numpy import *
import pylab as p
#import matplotlib.axes3d as p3
import mpl_toolkits.mplot3d.axes3d as p3
```

```python
fig = p.figure()
ax = p3.Axes3D(fig)

x,y = mgrid[0:100,0:100]
ax.plot_surface(y, x, data1, cmap=cm.Blues)
ax.plot_surface(y, x, dataN, cmap=cm.Reds)

ax.set_xlabel('M')
ax.set_ylabel('N')
ax.set_zlabel('total time')
#ax.axis([0,10,0,10])
p.show()
```

```python
for N in [1,99]:
    axis([0,10,0.0,0.10])
    plot(data1[:,N], 'b,-')
    plot(dataN[:,N], 'r,-')
    #savefig('plot-{0}.png'.format(M))
    figure()
```

```python
for M in [1,99]:
    axis([0,10,0.0,0.10])
    plot(data1[M,:], 'b,-')
    plot(dataN[M,:], 'r,-')
    #savefig('plot-{0}.png'.format(M))
    figure()
```
